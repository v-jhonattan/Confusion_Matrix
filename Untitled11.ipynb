```python
import numpy as np

def cosine_similarity_np(q, G):
    """
    q: (1, D)   G: (N, D)
    retorna vetor (N,) com similaridades cosseno
    """
    qn = q / (np.linalg.norm(q, axis=1, keepdims=True) + 1e-12)
    Gn = G / (np.linalg.norm(G, axis=1, keepdims=True) + 1e-12)
    return (qn @ Gn.T)[0]
```

```python
import os, glob, time, cv2, numpy as np, matplotlib.pyplot as plt
from mtcnn.mtcnn import MTCNN
from deepface import DeepFace

def imshow_bgr(title, img_bgr, size=10):
    plt.figure(figsize=(size, size*img_bgr.shape[0]/img_bgr.shape[1]))
    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.title(title); plt.show()

def draw_box_label(img, box, label):
    x,y,w,h = box
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
    cv2.putText(img, label, (x, max(0,y-6)), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

def safe_crop(img, x,y,w,h, m=0.20):
    H,W = img.shape[:2]; dx,dy = int(w*m), int(h*m)
    x1,y1 = max(0,x-dx), max(0,y-dy); x2,y2 = min(W,x+w+dx), min(H,y+h+dy)
    return img[y1:y2, x1:x2], (x1,y1,x2-x1,y2-y1)

detector = MTCNN()
def detect_faces_bgr(img_bgr, min_size=24, min_conf=0.70):
    rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    dets = detector.detect_faces(rgb)
    out=[]
    for d in dets:
        x,y,w,h = d['box']; conf=d['confidence']
        if w>=min_size and h>=min_size and conf>=min_conf:
            out.append({'box':(x,y,w,h),'confidence':conf})
    return out

try:
    EMB_MODEL = DeepFace.build_model("Facenet512")
except Exception:
    EMB_MODEL = None

def face_embedding(face_bgr):
    face_rgb = cv2.cvtColor(face_bgr, cv2.COLOR_BGR2RGB)
    try:
        reps = DeepFace.represent(img_path=face_rgb, model_name="Facenet512",
                                  detector_backend="skip", enforce_detection=False,
                                  model=EMB_MODEL)
    except TypeError:
        reps = DeepFace.represent(img_path=face_rgb, model_name="Facenet512",
                                  detector_backend="skip", enforce_detection=False)
    return np.array(reps[0]["embedding"], dtype=np.float32)

def cosine_similarity_np(q, G):
    qn = q / (np.linalg.norm(q, axis=1, keepdims=True) + 1e-12)
    Gn = G / (np.linalg.norm(G, axis=1, keepdims=True) + 1e-12)
    return (qn @ Gn.T)[0]

from collections import defaultdict
def load_gallery_embeddings(gallery_dir="/content/galeria"):
    id2embs = defaultdict(list)
    people = [p for p in sorted(glob.glob(os.path.join(gallery_dir,"*"))) if os.path.isdir(p)]
    for person_dir in people:
        person = os.path.basename(person_dir)
        for img_path in sorted(glob.glob(os.path.join(person_dir, "*"))):
            img = cv2.imread(img_path);
            if img is None: continue
            dets = detect_faces_bgr(img)
            if not dets: continue
            x,y,w,h = dets[0]['box']; crop,_ = safe_crop(img,x,y,w,h)
            try: id2embs[person].append(face_embedding(crop))
            except Exception as e: print(f"[ERR] {img_path}: {e}")
    gallery = {k: np.mean(np.vstack(v), axis=0) for k,v in id2embs.items() if v}
    print("IDs carregados:", list(gallery.keys()))
    return gallery

def recognize_with_gallery(img_bgr, gallery, threshold=0.45):
    dets = detect_faces_bgr(img_bgr)
    names = list(gallery.keys()); results=[]
    if not names:
        for d in dets:
            x,y,w,h = d['box']; _,box = safe_crop(img_bgr,x,y,w,h)
            results.append((box,"face",0.0))
        return results
    G = np.vstack([gallery[n] for n in names])
    for d in dets:
        x,y,w,h = d['box']; crop,box = safe_crop(img_bgr,x,y,w,h)
        q = face_embedding(crop).reshape(1,-1)
        sims = cosine_similarity_np(q, G)
        j = int(np.argmax(sims)); sim=float(sims[j])
        name = names[j] if sim>=threshold else "desconhecido"
        results.append((box,name,sim))
    return results
```

```python

cands=[]
for ext in ("jpg","jpeg","png"): cands += glob.glob(f"/content/**/*.{ext}", recursive=True)
for p in sorted(cands): print(" -", p)

# Definir o caminho EXATO da imagem
img_path_multi = "/content/galeria/pessoa_A/IMG_0109.jpeg"
img = cv2.imread(img_path_multi); assert img is not None, img_path_multi

# Visualizar índices (esquerda→direita)
dets = sorted(detect_faces_bgr(img), key=lambda d: d['box'][0])
print("Detectadas:", len(dets))
vis = img.copy()
for i,d in enumerate(dets):
    x,y,w,h = d['box']; _,box = safe_crop(img,x,y,w,h)
    draw_box_label(vis, box, f"#{i}")
imshow_bgr("Índices (esq→dir)", vis)



names4 = ["Graham O'Brien", "Doutor", "Ryan Sinclair", "Yasmin Khan"]

ts=int(time.time())
for i,d in enumerate(dets):
    if i>=len(names4): break
    x,y,w,h = d['box']; crop,_ = safe_crop(img,x,y,w,h)
    person_dir = f"/content/galeria/{names4[i]}"; os.makedirs(person_dir, exist_ok=True)
    out = f"{person_dir}/l2r_{ts}_{i}.jpg"
    cv2.imwrite(out, crop)
    print(f"[OK] {names4[i]} <- face #{i}: {out}")

gallery = load_gallery_embeddings("/content/galeria")
```

```python
import cv2

def recognize_and_show(img_path, gallery, threshold=0.45):
    img = cv2.imread(img_path)
    assert img is not None, f"Caminho inválido: {img_path}"
    assert len(gallery) > 0, "Galeria vazia. Refaça o Passo 1."

    res = recognize_with_gallery(img, gallery, threshold=threshold)
    vis = img.copy()
    for (box, name, score) in res:
        draw_box_label(vis, box, f"{name} ({score:.2f})")
    imshow_bgr("Detecção + Reconhecimento (Galeria)", vis)

recognize_and_show(img_path_multi, gallery, threshold=0.45)
```
